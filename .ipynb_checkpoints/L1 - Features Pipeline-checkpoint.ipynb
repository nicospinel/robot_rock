## Discover features

### Albums

```python
daft_albums = pd.read_csv('daft_albums.csv')
daft_albums
```

### Tracks

```python
tracks = []
for album in daft_albums['uri'].values:
    album_data = sp.album_tracks(album_id = album)
    for track in album_data['items']:
        track_features = sp.audio_features(track['uri'])[0]
        track_features['name'] = track['name']
        track_features['preview_url'] = track['preview_url']
        
        tracks.append(pd.DataFrame(track_features, index = [0]))
```

```python
daft_tracks = pd.concat(tracks, ignore_index = True)
daft_tracks.to_csv('daft_tracks.csv', index = False)
```
